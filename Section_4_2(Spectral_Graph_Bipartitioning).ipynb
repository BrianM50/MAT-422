```python
import matplotlib.pyplot as plt
from numpy import random
from sklearn.cluster import SpectralClustering
from sklearn.datasets import make_blobs

print('The aim of graph partition is to find partition such that the cut (the total number of edges between two disjoint sets of nodes) is minimized.')
```

```python
#Generate simple dataset
x, y = make_blobs(n_samples=500, centers=5, cluster_std=2)
plt.scatter(x[:,0], x[:,1])
plt.show()
```

```python
print('In machine learning and data analysis Special Clustering is a techniique used for clustering data points based similaritties between them.')
print('The SpectralClustering class applies the clustering to a projection of the normalized Laplacian')

#Define model using the SpeicalClustering class
sc = SpectralClustering(n_clusters=5).fit(x)
labels = sc.labels_
plt.scatter(x[:,0], x[:,1], c=labels)
plt.show()
```

```python
#Changes the number of clusters
f = plt.figure()
f.add_subplot(2, 2, 1)
for i in range(2, 6):
 sc = SpectralClustering(n_clusters=i).fit(x)
 f.add_subplot(2, 2, i-1)
 plt.scatter(x[:,0], x[:,1], s=5, c=sc.labels_, label="n_cluster-"+str(i))
 plt.legend()

plt.show()
```

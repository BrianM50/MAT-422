```python
import networkx as nx
import numpy as np

print('A graph consist of both a set of objects, called nodes, and the connections between these nodes, called edges.')
```

```python
#Example of a 3 regular graph with 6 nodes
G = nx.random_regular_graph(3, 6)
nx.draw(G)
```

```python
print('A matrix that represents one of these graphs is known as a Laplacian matrix.')
print('The Laplacian matrix of a simple graph G with n vertices is defined as:')
print('L = D - A, \n where D is the degree matrix and A is the adjacency matrix of the graph. \n')


A = nx.adjacency_matrix(G).toarray()
D = np.diag(np.ones(6) * 3)
L = D - A
print('The Laplacian matrix for the graph above is:')
print(L, '\n')
print('Characteristics of the matrix: ')
print('The diagonal contains the degrees of the nodes which is 3.')
print('A connection between the pairs of nodes is represented, where there is a -1.')
print('There is no connection between the pairs of nodes, where there is a 0')
```

```python
#This should be the same as the Laplacian matrix we calculated eariler
La = nx.laplacian_matrix(G).toarray()
print(La)
#We will check to see if they are the same
np.array_equal(nx.laplacian_matrix(G).toarray(), D - A)
```

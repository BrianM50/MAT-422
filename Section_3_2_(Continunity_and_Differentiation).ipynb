```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
print('Limits are the behaviors of a function near a certain point and are used to find continuity, derivatives, and integrals.')
print('A continuous function is a function that has no abrupt changes in value, meaning no breaks in the graph of the function.')
```

```python
print('Example 1: Find the limit of sin(x)/x as x approaches 0')
x = sy.Symbol('x')
f = sy.sin(x)/x
Lim = sy.limit(f,x,0)
print('Limit =', Lim)
print('As x approaches 0 the limit is 1 \n', 'This function is not continuous because at x = 0, y is undefined.')
```

```python
print('Example 2: Find the limit of x^2+3x+1 as x approaches infinity')
x = sy.Symbol('x')
g = x*x+3*x+1
lim = sy.limit(g,x,sy.oo)
print('Limit =', lim)
print('As x approaches 0 the limit is infinity. \n', 'This function is continuous because all x in the function are defined.')
```

```python
print('The derivative measures the sensitivity of change of a functions output with respect to the input.\n')
print('Example 1: Find the derivative of f(x) = x^3 + 2x^2 +2x')
x = sy.Symbol('x')
h = x*x*x+2*x*x+2*x
dif = sy.diff(h,x)
print('f\'(x) =', dif, '\n')

print('Example 2: Find the derivative of f(x) = xe^x')
x = sy.Symbol('x')
j = x*sy.exp(x)
dif = sy.diff(j,x)
print('f\'(x) =', dif)
```

```python
print('A partial derivative is a derivative of a function of multiple variables with respect to one variable, while the others are treated as constants. \n')

print('Example 3: Find the partial derivative of f(x) = x^4+x*y^4 with respect to x')
x , y = sy.symbols('x y')
f = x**4+x*y**4
dif = sy.diff(f,x)
print('f\'(x) =', dif, '\n')

print('Example 4: Find the partial derivative of f(x) = x^4+x*y^4 with respect to y')
x , y = sy.symbols('x y')
f = x**4+x*y**4
dif = sy.diff(f,y)
print('f\'(x) =', dif)
```

```python
import numpy as np
import matplotlib.pyplot as plt
print('Taylor\'s Theorem results in an approximation of a differentiable function around a given point by a polynomial.')
print('We will plot the function sin(x) and the first, third, fifth, seventh, and ninth order Taylor series approximations to see how close the approixamations we get are to the function sin(x).')

x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order', 'Ninth Order']

plt.figure(figsize = (10,8))
for n, label in zip(range(5), labels):
    y = y + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)
    plt.plot(x,y, label = label)

plt.plot(x, np.sin(x), 'k', label = 'Analytic')
plt.grid()
plt.title('Taylor Series Approximations of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
print('From the graph we see that the closest approximations to the function sin(x) are the seventh and ninth order. \n')
print('To compare the seventh and ninth order Taylor series approixmations, we will use sin(π/2) = 1.')

x = np.pi/2
y7 = 0
y9 = 0

for n in range(4):
    y7 = y7 + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)

for n in range(5):
    y9 = y9 + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)

print('Seventh order Taylor series approximation =',y7)
print('Ninth order Taylor series approximation =',y9)
print('We can see that both of these orders are really close to the actual value of the function.')
```

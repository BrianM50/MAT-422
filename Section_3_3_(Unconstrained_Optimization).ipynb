```python
print('The point x∗ ∈ R^d is a local minimizer of f over R^d if there is δ > 0 such that')
print('f(x) ≥ f(x∗), ∀x ∈ B δ(x∗)\{x∗} \n')

print('The point x∗ ∈ R^d is a global minimizer of f over R^d if')
print('f(x) ≥ f(x∗), ∀x ∈ R^d ')
```

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

x = np.arange(-2.5, 2, 0.01)
def f(x):
    return x**4 - 4*x**2 + 2*x - 1

grid = (-10, 10, 0.1)
xmin_global = optimize.brute(f, (grid, ))
print("Global minima = %s" % xmin_global)

xmin_local = optimize.fminbound(f, 0, 2)
print("Local minimum = %s" % xmin_local)

fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)
ax.plot(x, f(x), 'b-', label="f(x)")
ax.plot(xmin_global[0], f(xmin_global[0]), 'ro', label="Global Minima")
ax.plot(xmin_local, f(xmin_local), 'go', label="Local Minimum")
ax.legend(loc='best')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.axhline(0, color='black')
plt.show()
```

```python
print('The Gradient Descent is used to find a local minimum of a differentiable function by using an iterative optimization algorithm')

def gradient_descent(gradient, start, learn_rate, n_iter = 100, tolerance=1e-06):
  x = start
  plt.scatter(x, x**4 - 4*x**2 + 2*x - 1, c = 'red')
  for _ in range(n_iter):
    diff = -learn_rate * gradient(x)
    if np.all(np.abs(diff) <= tolerance):
      break
    x += diff
    plt.scatter(x, x**4 - 4*x**2 + 2*x - 1, c = 'red')
  plt.scatter(x, x**4 - 4*x**2 + 2*x - 1, c = 'blue')
  return x

x = np.linspace(-2.5,2,50)
y = x**4 - 4*x**2 + 2*x - 1
plt.plot(x, y, c = 'black')
print(gradient_descent(gradient=lambda x: 4*x**3 - 8*x + 2, start = 0, learn_rate = 0.09))
plt.show()
```
